Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.svm import SVC
```

Data Collection and Analyses

```python
# Loading The Diabetes Dataset Into Pandas DataFrame
diabetes_dataset = pd.read_csv('Diabetes_dataset.csv')
```

```python
# Printing the first five rows of the dataframe
diabetes_dataset.head()
```

```python
# Printing the last five rows of the dataframe
diabetes_dataset.tail()
```

```python
# Printing the number of rows and column in the dataframe
diabetes_dataset.shape
```

```python
# Printing the statistical measures of the dataframe
diabetes_dataset.describe()
```

```python
# Printing the number of diabetic and non-diabetic in the outcome column
diabetes_dataset['Outcome'].value_counts()
```

Non-diabetic ---> 0,
Diabetic ---> 1

```python
# Spliting the dataframe into features and target
features = diabetes_dataset.drop(columns ='Outcome', axis=1)
target = diabetes_dataset['Outcome']
X = features
Y = target
```

```python
# Printing the features and target
print(X)
print(Y)
```

```python
# Splitting dataset into training and test dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=5)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# The algorithm in use: Support vector Machine Classifier
classifier = SVC(kernel='linear')
```

```python
# Training the SVM Classifier
classifier.fit(X_train, Y_train)
```

MODEL EVALUATION: Using Accuracy Score

```python
# Accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score on training data : ', training_data_accuracy)
```

```python
# Accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on test data : ', test_data_accuracy)
```

BUILDING A SYSTEM CAPABLE OF PREDICTING DIABETES

```python
# Taking new data as input
input_data = (1,89,66,23,94,28.1,0.167,21)

# Changing the new data into numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping the array so prediction can be made for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardizing the imput data
scaler = StandardScaler() # Creating a standard scaler object
scaler.fit(input_data_reshaped)
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0]==0):
    print('The Person is Non-diabetic')
else:
    print('The Person is Diabetic')
```
